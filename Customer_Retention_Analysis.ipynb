**Data Cleaning and Transformation**

```python
# Convert 'Total Charges' to numeric, coerce errors to NaN
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors='coerce')
df
```

```python
#Fill missing 'Total Charges' with median
df['Total Charges'].fillna(df['Total Charges'].median(), inplace=True)
df
```

```python
# Encode binary categorical fields
binary_cols = ['Gender', 'Partner', 'Dependents', 'Paperless Billing', 'Senior Citizen']
for col in binary_cols:
    if df[col].dtype == 'object':
        df[col] = df[col].map({'No': 0, 'Yes': 1, 'Male': 0, 'Female': 1})
```

**EDA**

```python
# Outlier detection for Monthly Charges
plt.figure(figsize=(8,2))
sns.boxplot(x=df['Monthly Charges'])
plt.title('Monthly Charges Outlier Detection')
plt.show()
```

```python
# Calculate limits
Q1 = df['Monthly Charges'].quantile(0.25)
Q3 = df['Monthly Charges'].quantile(0.75)
IQR = Q3 - Q1
upper_limit = Q3 + 1.5 * IQR

print("Upper limit for Monthly Charges (outlier cap):", upper_limit)
print("Max Monthly Charges before capping:", df['Monthly Charges'].max())

# Cap outliers
df['Monthly Charges'] = np.where(df['Monthly Charges'] > upper_limit, upper_limit, df['Monthly Charges'])

print("Max Monthly Charges after capping:", df['Monthly Charges'].max())

# Visualize after capping
sns.boxplot(x=df['Monthly Charges'])
plt.title('Monthly Charges After Outlier Capping')
plt.show()
```

**EDA**

```python
# Churn Rate
churn_rate = df['Churn Label'].value_counts(normalize=True)
print("\nChurn Rate (%):\n", churn_rate * 100)
sns.countplot(x='Churn Label', data=df)
plt.title('Churn vs Non-Churn Distribution')
plt.show()
```

```python
# Tenure vs Churn
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Tenure Months', hue='Churn Label', multiple='stack', bins=20)
plt.title('Tenure Months Distribution by Churn')
plt.show()
```

```python
# Contract Type vs Churn
contract_churn = pd.crosstab(df['Contract'], df['Churn Label'], normalize='index')
print("\nContract Type vs Churn:\n", contract_churn)
contract_churn.plot(kind='bar', stacked=True)
plt.title('Churn Rate by Contract Type')
plt.ylabel('Proportion')
plt.show()
```

```python
# Payment Method vs Churn
payment_churn = pd.crosstab(df['Payment Method'], df['Churn Label'], normalize='index')
print("\nPayment Method vs Churn:\n", payment_churn)
payment_churn.plot(kind='bar', stacked=True)
plt.title('Churn Rate by Payment Method')
plt.ylabel('Proportion')
plt.show()
```

```python
# Churn Reasons (only for churned customers)
if 'Churn Reason' in df.columns:
    churned = df[df['Churn Label'] == 'Yes']
    print("\nNumber of churned customers:", churned.shape[0])
    print("Top 10 churn reasons:\n", churned['Churn Reason'].value_counts().head(10))
    plt.figure(figsize=(8,5))
    sns.countplot(y='Churn Reason', data=churned, order=churned['Churn Reason'].value_counts().index[:10])
    plt.title('Top 10 Churn Reasons')
    plt.xlabel('Count')
    plt.ylabel('Churn Reason')
    plt.show()
```

```python
# Services vs Churn
service_cols = ['Internet Service', 'Streaming TV', 'Streaming Movies']
for col in service_cols:
    if col in df.columns:
        service_churn = pd.crosstab(df[col], df['Churn Label'], normalize='index')
        print(f"\n{col} vs Churn:\n", service_churn)
        service_churn.plot(kind='bar', stacked=True)
        plt.title(f'Churn Rate by {col}')
        plt.ylabel('Proportion')
        plt.show()
```

```python
# Demographics vs Churn
demo_cols = ['Senior Citizen', 'Gender', 'Dependents']
for col in demo_cols:
    if col in df.columns:
        sns.countplot(x=col, hue='Churn Label', data=df)
        plt.title(f'Churn by {col}')
        plt.show()
```

```python
 #Charges vs Churn
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn Label', y='Monthly Charges', data=df)
plt.title('Monthly Charges by Churn')
plt.show()
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn Label', y='Total Charges', data=df)
plt.title('Total Charges by Churn')
plt.show()
```

```python
# Geographic distribution of churn (Top 10 churn cities)
if 'City' in df.columns:
    city_churn = df[df['Churn Label'] == 'Yes']['City'].value_counts().head(10)
    print("\nTop 10 Cities by Churn Count:\n", city_churn)
    plt.figure(figsize=(8,5))
    sns.barplot(y=city_churn.index, x=city_churn.values, orient='h')
    plt.title('Top 10 Cities by Churn Count')
    plt.xlabel('Churn Count')
    plt.ylabel('City')
    plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12,8))
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='YlGnBu')
plt.title('Feature Correlation Heatmap')
plt.show()
```
